```python
x.is_contiguous()
```

```python
contigious_tensor = torch.arange(0, 20, 1)
contigious_tensor = contigious_tensor.reshape(4, -1)
transpose_ct = contigious_tensor.T

print(f"\ncontigious_tensor:\n{contigious_tensor}")
print(f"transpose_ct:\n{transpose_ct}")

print(f"\ncontigious_tensor.is_contiguous: {contigious_tensor.is_contiguous()}")
print(f"transpose_ct.is_contiguous: {transpose_ct.is_contiguous()}")

print(f"\n\ncontigious_tensor.view(5, -1): \n{contigious_tensor.view(5, -1)}")

try:
    print(f"transpose_ct.view(5, -1): \n{transpose_ct.view(4, -1)}")
except RuntimeError:
    print("transpose_ct.view will cause an error")
```

### Squeeze & unsqueeze

```python
x
```

```python
print(f"x:\n{x}")
print(f"x.shape: {x.shape}")
print(f"x.ndim: {x.ndim}")
```

```python
squeezed_x = x.squeeze()
print(f"squeezed_x:\n{squeezed_x}")
print(f"squeezed_x.shape: {squeezed_x.shape}")
print(f"squeezed_x.ndim: {squeezed_x.ndim}")
```

```python
x = torch.tensor([[1, 2, 3]])
print(f"x:\n{x}")
print(f"x.shape: {x.shape}")
print(f"x.ndim: {x.ndim}\n\n")

squeezed_x = x.squeeze()
print(f"squeezed_x:\n{squeezed_x}")
print(f"squeezed_x.shape: {squeezed_x.shape}")
print(f"squeezed_x.ndim: {squeezed_x.ndim}")
```

```python
x = torch.tensor([[1, 2, 3], [4, 5, 6]])
print(f"x:\n{x}")
print(f"x.shape: {x.shape}")
print(f"x.ndim: {x.ndim}\n\n")

squeezed_x = x.unsqueeze(dim = 0)
print(f"squeezed_x:\n{squeezed_x}")
print(f"squeezed_x.shape: {squeezed_x.shape}")
print(f"squeezed_x.ndim: {squeezed_x.ndim}")
```

### Permute

```python
x = torch.rand(size=(224, 224, 3))
```

```python
print(f"x.shape: {x.shape}")
print(f"x.ndim: {x.ndim}\n\n")
print(f"x:\n{x}")
```

```python
x_permuted = x.permute(2, 0, 1) # axis 0 -> 1, 1 -> 2, 2->0
print(f"x_permuted.shape: {x_permuted.shape}")
print(f"x_permuted.ndim: {x_permuted.ndim}\n\n")
print(f"x_permuted:\n{x_permuted}")
```
