```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
firstData = np.linspace(0, 10, 20)
```

```python
firstData
```

```python
secondData = firstData ** 2
```

```python
secondData
```

```python
figures, axeses = plt.subplots()
axeses.plot(firstData, secondData, "--", color="#cafb02", alpha=1)
axeses.plot(secondData, firstData, "--", color="#ffcd07")
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()

new_axes.plot(
    firstData, firstData + 2, 
    color="blue", linewidth=1.0,
    marker=".", markersize= 6,
)
new_axes.plot(
    firstData, firstData + 4, 
    color="yellow", linewidth=3.0, linestyle=":",
)
```

```python
plt.boxplot(firstData)
plt.show()
```

```python
plt.scatter(firstData, secondData)
plt.show()
```
