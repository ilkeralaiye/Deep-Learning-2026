```python
age_list = np.random.randint(0, 80, 10)
```

```python
age_list.sort()
```

```python
weight_list = np.random.randint(50, 100, 10)
```

```python
weight_list.sort()
```

```python
plt.plot(age_list, weight_list, "r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age - Weight")
plt.show()
#plt.savefig("data.png")
```

```python
linear_array = np.linspace(-20,+20,39)
```

```python
linear_array
```

```python
linear_array_square = linear_array**2
```

```python
plt.plot(linear_array, linear_array_square, "--")
plt.xlabel("Number")
plt.ylabel("Square of number")
plt.title("Square of given numbers")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(linear_array, linear_array_square, "g--")
plt.subplot(1, 2, 2)
plt.plot(age_list, weight_list, "r--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.5,0.5])
figure_axes.plot(linear_array, linear_array_square, "g--")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Figure Image")
```

```python
new_fig = plt.figure(dpi=80)
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(linear_array, linear_array**2, "g--", label="numpy array square")
new_axes.plot(linear_array, linear_array * 10, "r--", label="numpy array * 10")
new_axes.legend()
plt.show()
```
