```python
x_tensor = torch.tensor([1, 2, 3])
x_tensor
```

```python
print("Arithmetical Operations")
print(f"x_tensor + 5 = {x_tensor + 5}")
print(f"x_tensor - 5 = {x_tensor - 5}")
print(f"x_tensor * 5 = {x_tensor * 5}")
print(f"x_tensor / 5 = {x_tensor / 5}")
```

```python
print("Arithmetical Operations with Functions")
print(f"x_tensor + 5 = {x_tensor.add(5)}")
print(f"x_tensor - 5 = {x_tensor.subtract(5)}")
print(f"x_tensor * 5 = {x_tensor.multiply(5)}")
print(f"x_tensor / 5 = {x_tensor.divide(5)}")
```

```python
print("Arithmetical Operations between tensors")
print(f"x_tensor + x_tensor = {x_tensor.add(x_tensor)}")
print(f"x_tensor - x_tensor = {x_tensor.subtract(x_tensor)}")
print(f"x_tensor * x_tensor = {x_tensor.multiply(x_tensor)}")
print(f"x_tensor / x_tensor = {x_tensor.divide(x_tensor)}")
```
