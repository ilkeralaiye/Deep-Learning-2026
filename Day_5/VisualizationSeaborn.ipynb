```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("athlete_events.csv")
```

```python
data.columns
```

```python
data.describe()
```

```python
data.head()
```

```python
data[data["Medal"].notna()].describe()
```

```python
plt.scatter(x="Height", y="Weight", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs Weight Table")
plt.show()
```

# Seaborn

```python
sns.set_style("darkgrid")
sns.scatterplot(x="Height", y="Weight", hue="Sex", data=data)
plt.xlabel("Height of Atheletes")
plt.xlabel("Weight of Atheletes")
plt.show()
```

```python
data.columns
```

```python
data["Medal"].unique()
```

```python
atheletes_with_medal = data[data["Medal"].notna()]

sns.set_style("darkgrid")
sns.scatterplot(x="Height", y="Weight", hue="Sex", size="Medal", data=atheletes_with_medal)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.show()
```

```python
sns.lineplot(x="Height", y="Weight", hue="Sex", data=atheletes_with_medal)
plt.xlabel("Height of Atheletes")
plt.ylabel("Weight of Atheletes")
plt.show()
```

```python
sns.set_style("dark")
sns.displot(x="Height", hue="Sex", data=data)
plt.ylabel("Frequency")
plt.title("Atheletes Height Distribution")
plt.show()
```

```python
sns.displot(x="Year", hue="Sex", data=data, kind="kde")
plt.ylabel("Frequency")
plt.title("Atheletes Medal Distribution")
plt.show()
```

```python
sns.barplot(x="Medal",
            y="Height",
            hue="Sex",
            data=data)
plt.title("Medals by height")
plt.show()
```

```python
sns.catplot(x="Medal", y="Height", hue="Sex", col="Season", data=data)
```
