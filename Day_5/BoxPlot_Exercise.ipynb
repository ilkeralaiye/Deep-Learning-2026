```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# BoxPlot
* Median
* Quartile
* Min, Max (1.5 * IQR)
* Outlier

```python
data = np.array(np.random.randint(0, 100, 50))
```

```python
plt.figure(figsize = (5, 5))
plt.title("BoxPlot")
plt.ylabel("Numbers")
plt.grid(True) 

sns.boxplot(y = data)

plt.show()
```

```python
titanic_data = sns.load_dataset(name="titanic")
```

```python
titanic_data
```

```python
plt.figure(figsize=(10, 5))

# First Graph
plt.subplot(1, 2, 1)
sns.boxplot(x="class", y="age", data=titanic_data)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

# Second Graph
plt.subplot(1, 2, 2)
sns.boxplot(x="class", y="fare", data=titanic_data)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.show()
```
